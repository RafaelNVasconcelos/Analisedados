```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fish_consumption = pd.read_csv('fish-and-seafood-consumption-per-capita.csv')
```

```python
fish_consumption.head()
```

```python
plt.figure(figsize=(14,7))

sns.lineplot(x=fish_consumption['Year'],y=fish_consumption['Fish, Seafood- Food supply quantity (kg/capita/yr) (FAO, 2020)'])
```

```python
#filtro Brasil
fish_consumption_br = fish_consumption[fish_consumption['Code']=='BRA']
```

```python
fish_consumption_br.tail()
```

```python
plt.figure(figsize=(14,7))
sns.lineplot(x=fish_consumption_br['Year'],y=fish_consumption_br['Fish, Seafood- Food supply quantity (kg/capita/yr) (FAO, 2020)'])
plt.title("Comsumo de peixes no Brasil ao longo dos anos",size=16)
```
